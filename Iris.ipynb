# Label Encoding

```python
le= LabelEncoder()
```

```python
df['Species']= le.fit_transform(df['Species'])
df.head()
```

# Model Training

```python
#splitting data into training and testing. Train- 70% Test- 30%
x=df.drop(columns=["Species"])
y=df["Species"]
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.30, shuffle= True)
```

```python
#Logistic Regression
model= LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Printing Metric
print("accuracy is equal to ", model.score(x_test, y_test)* 100)
```

```python
#KNN
model = KNeighborsClassifier()
model.fit(x_train, y_train)
```

```python
#Printing Metric
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# decision tree
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
#Printing Metric
print("Accuracy: ",model.score(x_test, y_test) * 100)
```
